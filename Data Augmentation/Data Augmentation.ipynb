```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

###### Both behave like OPenCV and helps to work on images

```python
img = image.load_img('Image/image1.jpg' , target_size = (200,200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
type(img) # image is not a numpy array. We cannot directly apply array operations on image
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30, # angle at which you want to rotate your image
    shear_range = 0.2, # distort image at 0.2
    zoom_range = 0.2, # it could be zoom in or zoom out -- > 0.2 is maximum is k and andr value rhy gy
    horizontal_flip = True, # image ko flip kr de ga left right change ho ga
    #vertical_flip = True, --> vertically flip ho ga .... top, bottom change ho ga ... is ki zroort abhi nhi hai kyu k image ko ulta kr de ga
    width_shift_range = 0.2, # thora sa image ko apni jaga se hilaye ga
    height_shift_range = 0.2) # thora sa image ko apni jaga se hilaye ga

## fill_mode = 'nearest' --> ye image k shift hone prr jo pixels khali ho jaty hain ye un ko fill krwata hai
##### nearest -- > aas pass wale pixels se fill kre ga
##### constant --> pixels ko black kr de ga
##### reflection --> ooper wale pixels se fill kre ga
```

###### In above block of code we defined the image transformations

### Converting the image into numpy array

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3) # imgae ko batch ki form mein dena hota hai is lye reshape kr k ye define kya hai k hmm 1 hi image ko process krein gy
```

# Genaerating the images

```python
i = 0
for output in datagen.flow(input_batch,batch_size = 1,save_to_dir = 'aug'):
    i = i + 1

    if i == 10:
        break
```

###### flow --> hmm single image pr kaam kr rhy hain is lye flow ka use kya hai 
###### batch_size = 1 -- > output mein aik batch form ho ga jis mein 1 image ho gy
###### save_to_dir = 'aug' --> naye generate hone wali image is folder mein store ho gy
