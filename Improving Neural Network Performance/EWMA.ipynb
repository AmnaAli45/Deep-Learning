```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("mean_temperature_data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
plt.scatter(df["Date"],df["Mean_Temp"])
plt.show()
```

```python
x1 = df["Mean_Temp"].ewm(alpha = 0.9).mean() # alpha = 1 - beta ---> beta = 1- alpha 
```

```python
x1
```

```python
df["EWMA"] = x1
```

```python
df
```

```python
plt.figure(figsize = (15,12))
plt.scatter(df["Date"],df["Mean_Temp"])
plt.plot(df["Date"],df["EWMA"] ,color = 'black')
plt.show()
```
