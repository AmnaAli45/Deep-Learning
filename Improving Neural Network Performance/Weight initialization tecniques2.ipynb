```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('binary_classification_dataset.csv')
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df ['label']
```

```python
plt.scatter(df['feature1'],df['feature2'],c=y)
plt.show()
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(10,activation='tanh',input_dim = 2))
model.add(Dense(10,activation='tanh'))
model.add(Dense(10,activation='tanh'))
model.add(Dense(10,activation='tanh'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
initial_weghts = model.get_weights()
initial_weghts[0] = np.random.randn(2,10)*np.sqrt(1/2)
initial_weghts[1] = np.zeros(model.get_weights()[1].shape) # setting biases to zero
initial_weghts[2] = np.random.randn(10,10)*np.sqrt(1/10)
initial_weghts[3] = np.zeros(model.get_weights()[3].shape)
initial_weghts[4] = np.random.randn(10,10)*np.sqrt(1/10)
initial_weghts[5] = np.zeros(model.get_weights()[5].shape)
initial_weghts[6] = np.random.randn(10,10)*np.sqrt(1/10)
initial_weghts[7] = np.zeros(model.get_weights()[7].shape)
initial_weghts[8] = np.random.randn(10,1)*np.sqrt(1/10)
initial_weghts[9] = np.zeros(model.get_weights()[9].shape)
```

```python
model.set_weights(initial_weghts)
```

```python
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
history = model.fit(x,y,epochs = 100,validation_split=0.2)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.values,y.values.astype(int),clf = model)
```

```python
loss,accuracy = model.evaluate(x,y)
print("loss: ",loss)
print("accuracy: ",accuracy)
```
